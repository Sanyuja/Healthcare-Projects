```python
train= df.drop(["SurgDate", "DOW","Actual", "T - 1","T - 2"], axis=1)

test = df[["T - 3","T - 2", "T - 1","Actual"]]
```

### ARMA (1,0) model

```python
Arma = sm.tsa.ARMA(train.iloc[50], order=(1,0))
model_results = Arma.fit()
(model_results.forecast(4)[0])[3]
(model_results.forecast(4))
```

```python
forecasts04=[]
conf_int04_up = []
conf_int04_ll = []
std_err04 = []
for x in range(0,len(train)):
    Arma = sm.tsa.ARMA(train.iloc[x], order=(2,0))
    model_results = Arma.fit()
    forecasts04.append((model_results.forecast(4)[0])[3])
    conf_int04_up.append((model_results.forecast()[2])[0][1])
    conf_int04_ll.append((model_results.forecast()[2])[0][0])
    std_err04.append(model_results.forecast()[1])
```

```python
arma04 = pd.DataFrame(columns=[ "SurgDate","DOW","Conf_int_up","Conf_int_ll","std_err", "forecasts","Actual"])
arma04["SurgDate"] = df["SurgDate"]
arma04["DOW"] = df["DOW"]
arma04["forecasts"]= (forecasts04)
arma04["Conf_int_up"]= (conf_int04_up)
arma04["Conf_int_ll"]= (conf_int04_ll)
arma04["forecasts"]= np.ceil(arma04["forecasts"])
arma04["Conf_int_up"]= np.ceil(arma04["Conf_int_up"])
arma04["Conf_int_ll"]= np.ceil(arma04["Conf_int_ll"])
arma04["std_err"]= std_err04
arma04["Actual"]= df["Actual"]
```

```python
plt.plot(arma04["SurgDate"], arma04["Actual"], label = 'Actual')
plt.plot(arma04["SurgDate"],arma04["forecasts"], label = 'Forcasts')
#plt.plot(arma04["SurgDate"],arma04["Conf_int_up"])
#plt.plot(arma04["SurgDate"],arma04["Conf_int_ll"])
plt.fill_between(arma04["SurgDate"], arma04["Conf_int_ll"], arma04["Conf_int_up"], alpha= 0.2, label= "Confidence Interval")
plt.legend()
plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=20)    
plt.rc('ytick', labelsize=20)  
plt.title('Forecasts', fontsize=20)
plt.ylabel('Number of surgeries on a day') 
plt.savefig("arma04")
```
